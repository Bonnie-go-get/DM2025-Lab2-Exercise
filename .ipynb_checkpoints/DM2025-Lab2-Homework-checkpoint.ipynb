#### Apply some further text cleaning, oversampling, and class weight

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
```

```python
def cleaned_text(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\s+", " ", text).strip()
    return text
train_df['cleaned_text'] = train_df['text'].apply(cleaned_text)
test_df['cleaned_text'] = test_df['text'].apply(cleaned_text)
```

```python
from sklearn.utils import resample
def oversample_dataframe(df, label_col="emotion"):
    df_list = []
    max_size = df[label_col].value_counts().max()
    for label in df[label_col].unique():
        df_label = df[df[label_col] == label]
        df_upsampled = resample(df_label, replace=True, n_samples=max_size, random_state=42)
        df_list.append(df_upsampled)
    df_oversampled = pd.concat(df_list).sample(frac=1, random_state=42)  # shuffle
    return df_oversampled

train_df_oversampled = oversample_dataframe(train_df, label_col="emotion")

print("Before oversampling:")
print(train_df['emotion'].value_counts())

print("\nAfter oversampling:")
print(train_df_oversampled['emotion'].value_counts())
```

# 2. Feature Engineering Steps

##### Firstly, I use the 'text' in train_df data, with BOW and TF-IDF for feature engineering and try on different models to see each model performance.
##### After that, I try on further prepocessing, imbalance handling, and other feature engineering to apply on models to see if these further changing can help improve the performance or not.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

import warnings
warnings.filterwarnings("ignore")
```

```python
X = train_df['text']
y = train_df['emotion']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
results = []
```

```python
def evaluate_model(name, model, X_test, y_test):
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)

    print(f"\n {name}")
    print("Accuracy:", acc)
    print(classification_report(y_test, preds))

    results.append([name, acc])
```

```python
# Bag-of-Words
bow_vectorizer = CountVectorizer(max_features=20000)
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
# TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=20000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

## 3. Model Implementation Steps

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
models = {
"Logistic Regression": LogisticRegression(max_iter=500),
"Naive Bayes": MultinomialNB(),
"Linear SVC": LinearSVC(),
"Random Forest": RandomForestClassifier(n_estimators=150),
"Gradient Boosting": GradientBoostingClassifier()
}
```

```python
for name, model in models.items():
    print("\nTraining", name, "with BOW")
    model.fit(X_train_bow, y_train)
    evaluate_model(name + " (BoW)", model, X_test_bow, y_test)
```

```python
for name, model in models.items():
    print("\nTraining", name, "with TF-IDF")
    model.fit(X_train_tfidf, y_train)
    evaluate_model(name + " (TF-IDF)", model, X_test_tfidf, y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.utils import to_categorical
```

```python
# Sequence preparation
tokenizer = Tokenizer(num_words=20000)
tokenizer.fit_on_texts(X_train)

X_train_seq = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=50)
X_test_seq = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=50)

emotion_labels = {e: i for i, e in enumerate(sorted(train_df["emotion"].unique()))}
y_train_ids = np.array([emotion_labels[e] for e in y_train])
y_test_ids = np.array([emotion_labels[e] for e in y_test])

num_classes = len(emotion_labels)
```

```python
# CNN model
cnn = Sequential([Embedding(20000, 128, input_length=50), Conv1D(128, 5, activation='relu'), GlobalMaxPooling1D(), Dense(64, activation='relu'), Dense(num_classes, activation="softmax")])
cnn.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
cnn.fit(X_train_seq, y_train_ids, epochs=3, batch_size=64, verbose=1)

cnn_preds = np.argmax(cnn.predict(X_test_seq), axis=1)
acc = accuracy_score(y_test_ids, cnn_preds)
print("\n Result of CNN model")
print("Accuracy:", acc)
print(classification_report(y_test_ids, cnn_preds))
results.append(["CNN", acc])
```

```python
# BiLSTM model
lstm = Sequential([Embedding(20000, 128, input_length=50), Bidirectional(LSTM(128)), Dense(64, activation='relu'), Dense(num_classes, activation='softmax')])
lstm.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
lstm.fit(X_train_seq, y_train_ids, epochs=3, batch_size=64, verbose=1)

lstm_preds = np.argmax(lstm.predict(X_test_seq), axis=1)
acc = accuracy_score(y_test_ids, lstm_preds)
print("\n Result of BiLSTM model")
print("Accuracy:", acc)
print(classification_report(y_test_ids, lstm_preds))
results.append(["BiLSTM", acc])
```

```python
summary = pd.DataFrame(results, columns=["Model", "Accuracy"])
print("\n Model comparison table")
print(summary.sort_values("Accuracy", ascending=False))
```

```python
# Predict the test_df by the most accuracy model and save into csv file
# Based on the above summary, I choose the TF-IDF and Logistic Regression combination
train_df = merged_df[merged_df['split'] == 'train'].copy()
test_df = merged_df[merged_df['split'] == 'test'].copy()

tfidf_vectorizer = TfidfVectorizer(max_features=20000)
X_train_tfidf = tfidf_vectorizer.fit_transform(train_df['text'])
X_test_tfidf = tfidf_vectorizer.transform(test_df['text'])
y_train = train_df['emotion']

log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train_tfidf, y_train)

test_predictions = log_reg.predict(X_test_tfidf)
submission_df = pd.DataFrame({'id': test_df['id'], 'emotion': test_predictions})
submission_df.to_csv('Lab2_Phase3/submission_1.csv', index=False)
print("Submission file 'submission_1.csv' created successfully.")
print(submission_df.head())
```

#### I try to see how the models work if including class_weight to handle the imbalance in the cleaned text of train_df

```python
X = train_df['cleaned_text']
y = train_df['emotion']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.utils.class_weight import compute_class_weight
```

```python
classes = np.unique(y_train)
class_weights_dict = dict(zip(classes, compute_class_weight(class_weight='balanced', classes=classes, y=y_train)))
print("Class weights:", class_weights_dict)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=500, class_weight=class_weights_dict),
    "Linear SVC": LinearSVC(class_weight=class_weights_dict),
    "Random Forest": RandomForestClassifier(n_estimators=150, class_weight=class_weights_dict),
}
```

```python
 results_cleaned_classweight = []
```

```python
def evaluate_model(name, model, X_test, y_test):
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)

    print(f"\n{name}")
    print("Accuracy:", acc)
    print(classification_report(y_test, preds))

    results_cleaned_classweight.append([name, acc])
```

```python
bow_vectorizer = CountVectorizer(max_features=20000)
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=20000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
for name, model in models.items():
    print("\nTraining", name, "with BOW")
    model.fit(X_train_bow, y_train)
    evaluate_model(name + " (BoW)", model, X_test_bow, y_test)
```

```python
for name, model in models.items():
    print("\nTraining", name, "with TF-IDF")
    model.fit(X_train_tfidf, y_train)
    evaluate_model(name + " (TF-IDF)", model, X_test_tfidf, y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
```

```python
cnn_class_weights = {i: w for i, w in enumerate([class_weights_dict[c] for c in sorted(classes)])}
```

```python
tokenizer = Tokenizer(num_words=20000)
tokenizer.fit_on_texts(X_train)

X_train_seq = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=50)
X_test_seq = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=50)

emotion_labels = {e: i for i, e in enumerate(sorted(train_df["emotion"].unique()))}
y_train_ids = np.array([emotion_labels[e] for e in y_train])
y_test_ids = np.array([emotion_labels[e] for e in y_test])

num_classes = len(emotion_labels)

cnn = Sequential([Embedding(input_dim=20000, output_dim=128, input_length=50), Conv1D(128, 5, activation='relu'), GlobalMaxPooling1D(), Dense(64, activation='relu'), Dense(num_classes, activation="softmax")])
cnn.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
cnn.fit(X_train_seq, y_train_ids, epochs=3, batch_size=64, verbose=1, class_weight=cnn_class_weights)
cnn_preds = np.argmax(cnn.predict(X_test_seq), axis=1)
acc = accuracy_score(y_test_ids, cnn_preds)
print("\nResult of CNN model")
print("Accuracy:", acc)
print(classification_report(y_test_ids, cnn_preds))
results_cleaned_classweight.append(["CNN", acc])
```

```python
summary = pd.DataFrame(results_cleaned_classweight, columns=["Model", "Accuracy"])
print("\n Model comparison table on train_df cleaned text with class_weight computed")
print(summary.sort_values("Accuracy", ascending=False))
```
